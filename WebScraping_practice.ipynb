```python
def verify_request(response):
    resp = response.status_code
    response_dict = {100:'Continue', 101:'Switching Protocols', 103:'Early Hints', 200:'OK', 201:'Created', 202:'Accepted', 203:'NAI',
                 204:'Reset Content', 206:'Partial Content', 300:'Multiple Choices', 301:'Moved Permanently', 302:'Found', 303:'See Other',
                 304:'Not Modified', 307:'Temporary Redirect', 308:'Permanent Redirect', 400:'Bad Request', 401: 'Unauthorized',
                 402:'Payment Required', 403:'Forbidden', 404:'Not Found', 405:'Method not allowed', 406:'Not Acceptable', 407:'Proxy Authentification Required',
                 408:'Request Timeout', 409:'Conflict', 410:'Gone', 411:'Length Required', 412:'Precondition Failed', 413:'Request Too Large', 414:'Request-URI Too Long',
                 415:'Unsupported Media Length', 416:'Range Not Satisfiable', 417:'Expectation Failed', 500:'Internal Server Error', 501:'Not Implemented',
                 502:'Bad Gateway', 503:'Service Not Availabe', 504:'Gateway Timeout', 505:'HTTP Version Not Supported', 511:'NA Required', 1020:'ahref'}
    if(resp<200):
        print("Gateway response type: Information", end="\n")
    elif(resp<300):
        print("Gateway response type: Successful", end="\n")
    elif(resp<400):
        print("Gateway response type: Redirection", end="\n")
    elif(resp<500):
        print("Gateway response type: Client Error", end="\n")
    else:
        print("Gateway response type: Server Error", end="\n")

    print(f'Details - Request type: {response_dict[resp]}')
```

```python
site_3 = 'https://books.toscrape.com/index.html'
response_3 = requests.get(site_3)

verify_request(response_3)
```

```python
soup_3 = BeautifulSoup(response_3.text, 'html.parser')
```

```python
links_ctg=[]
for i in soup_3.find('ul', class_= 'nav nav-list').findAll('a'):
  x = i.get('href')
  if x =='catalogue/category/books_1/index.html':
    pass
  else:
    links_ctg.append(x)
```

```python
first_half = 'https://books.toscrape.com/'
links_ctg = [first_half + i for i in links_ctg]
```

```python
links_ctg
```

```python
name = []
availability = []
category = []

for site_t in links_ctg:
  response_t = requests.get(site_t)
  soup_t= BeautifulSoup(response_t.text, 'html.parser')
  
  
  for i in soup_t.findAll('h3'):
    x = i.text
    name.append(x)
    
    z = soup_t.find('h1').text
    category.append(z)

  for i in soup_t.findAll('p', class_='instock availability'):
      y = i.text.replace(' ','').replace('\n','')
      availability. append(y)
   
  a = 1
  while soup_t.find('li',class_='next') != None:
    # if soup_t.find('li',class_='next') == None:
    #   print("It's not existent ")
    # else:
    #   print('exist')
    #site_t = site_t.replace('index','page-{}'.format(a))
    
    if a <=1:
      site_t = site_t.replace('index','page-1')
      response_t = requests.get(site_t)
      soup_t = BeautifulSoup(response_t.text, 'html.parser')
  
    else:
       
      site_t = site_t.replace('page-{}'.format(a-1),'page-{}'.format(a))
      response_t = requests.get(site_t)
      soup_t = BeautifulSoup(response_t.text, 'html.parser')
      
      z = soup_t.find('h1').text

      for i in soup_t.findAll('h3'):
        x = i.text
        name.append(x)
        category.append(z)
        
      for i in soup_t.findAll('p', class_='instock availability'):
        y = i.text.replace(' ','').replace('\n','')
        availability.append(y)
    a += 1 

    
```
