```python
## To make our life easier, we can group these into sensible lists and use those as keys

base_stats = ['name','total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
training = ['catch_rate', 'base_friendship', 'base_experience', 'growth_rate']
breeding = ['egg_type_number', 'egg_type_1','egg_type_2', 'percentage_male', 'egg_cycles',]
```

```python
## When using boolean masks, we can use logical operators inside the brackets

## OR 
df[(df['hp'] == 1) | (df['type_1'] == 'Bug')][['name', 'hp', 'type_1']]
```

```python
## AND 
df[(df['hp'] == 1) & (df['type_1'] == 'Bug')][['name', 'hp', 'type_1']]
```

```python
## Some pokemons appear multiple times on the list – different variations of the same pokemon. We can find these by the pokedex numbers
## First, we create a dataframe of the unique pokedex numbers and the count of each of these numbers

duplicate_pokemon = df.groupby('pokedex_number')[['name']].count().sort_values(['name'], ascending = False).reset_index()
```

```python
duplicate_pokemon.head()
```

```python
## Next, we retain only those entries that appeared multiple times

duplicate_pokemon = duplicate_pokemon[duplicate_pokemon['name'] > 1].copy()
```

```python
duplicate_pokemon.tail()
```

```python
## We can look at the results, but we only see the pokedex number and the count ('name')... not very useful

duplicate_pokemon.sort_values('pokedex_number')
```

```python
## We can join the two tables together using '.merge()'

new_df = df.merge(duplicate_pokemon, how = 'inner', on = 'pokedex_number')
new_df.rename(columns = {'name_y':'poke_count'}, inplace = True)
```

```python
new_df
```

```python
## We can visualise the count of pokemons by their type

sns.countplot(x='type_1',
              data=df,
              palette = 'magma',
              order = df['type_1'].value_counts().index )

plt.xticks(rotation=45,
           horizontalalignment='right',
           fontweight ='light',
           fontsize ='large'
           )

plt.title('Pokemon Count By Primary Type')

plt.savefig('plot.png', dpi = 300, bbox_inches = 'tight') # saving the image to the working directory

plt.show()
```

```python
## Create a chart of the pokemon heights ! 
## Who is tallest. Who is shortest?!
## Can you save your charts as image files?
df.head()
```

```python
#create a new dataframe and sort it from hightest to lowest
df_height = df[['name','height_m']].sort_values(by = 'height_m',ascending = False)
```

```python
#plot the 20 tallest pokemon

plt.figure(figsize = (20,10))
sns.barplot(data = df_height.head(20),
            x = 'name',
            y = 'height_m',
            palette = 'copper')

#adjust  x label
plt.xticks (rotation = 45,
            horizontalalignment = 'right',
            fontweight = 'light',
            fontsize = 'large')

#give title and change the name of labels
plt.title('20 Tallest Pokemon')
plt.xlabel('pokemon name')
plt.ylabel('height')

#save
plt.savefig('tallestPokemon.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
#plot the 20 shortest pokemon

#plot the 20 shortest
plt.figure(figsize = (20,5))

sns.barplot(data = df_height.tail(20),
            x = 'name',
            y = 'height_m',
            palette = 'rainbow')

#adjust x label
plt.xticks (rotation = 45,
            horizontalalignment = 'right',
            fontweight = 'heavy',
            fontsize = 'large')

#change labels and add the title
plt.title('20 Shortest Pokemons')
plt.xlabel('Pokemon Name')
plt.ylabel('Height')

#save

plt.savefig('ShortestPokemon.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
